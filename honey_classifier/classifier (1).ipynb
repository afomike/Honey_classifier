```python
columns = ['TorF', 'group', 'bot1', 'bot2', 'region', 'Al', 'B', 'Ba', 'Ca', 'Fe',
       'K', 'Mg', 'Mn', 'Na', 'P', 'Sr', 'Zn']
```

```python
columns_to_encode = ['TorF','group','bot1','bot2']
```

```python
df.head()
```

```python
corrmat =df.corr()
top_corr_features=corrmat.index
plt.figure(figsize=(10,10))
g=sns.heatmap(df[top_corr_features].corr(),annot=True, cmap="RdYlGn")
```

```python
# Drop specified columns and perform one-hot encoding
X = pd.get_dummies(df.drop(['region'], axis=1), columns=columns_to_encode)
# Extract the target variable
y = df['region']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rfe = RandomForestClassifier(max_depth=100, min_samples_split=18, min_samples_leaf=1)
rfe.fit(X_train, y_train)
y_pred= rfe.predict(X_test)
rfe_accuracy= accuracy_score(y_pred, y_test)
# Evaluating the Random Forest Model
print ("Accuracy on test set is ", round(rfe.score(X_test, y_test) * 100, 2), "%")
print ("Accuracy on train set is ", round(rfe.score(X_train, y_train) * 100, 2), "%")
```

```python
y_pred
```

```python
# Training the Decision Tree Classification model on the Training set
dec = DecisionTreeClassifier(max_depth=100, min_samples_split=18, min_samples_leaf=1)
dec.fit(X_train, y_train)
y_pred= dec.predict(X_test)
# Evaluating the Decision Tree Classifier
print ("Accuracy on Decision Tree Classifier test set is ", round(dec.score(X_test, y_test) * 100, 2), "%")
print ("Accuracy on Decision Tree Classifier train set is ", round(dec.score(X_train, y_train) * 100, 2), "%")
# cm = confusion_matrix(y_test, y_pred)
# print(cm)
```

```python
svc = SVC(C=0.1, kernel='linear')
svc.fit(X_train, y_train)
y_pred= svc.predict(X_test)
# Evaluating the SVM
print ("Accuracy on Support Vector Machine test set is ", round(svc.score(X_test, y_test) * 100, 2), "%")
print ("Accuracy on Support Vector Machine train set is ", round(svc.score(X_train, y_train) * 100, 2), "%")
# cm = confusion_matrix(y_test, y_pred)
# print(cm)
```

```python
naive = GaussianNB()
naive.fit(X_train, y_train)
y_pred = naive.predict(X_test)
# Evaluating the naive Model
print ("Accuracy on naive base Model test set is ", round(naive.score(X_test, y_test) * 100, 2), "%")
print ("Accuracy on naive base train set is ", round(naive.score(X_train, y_train) * 100, 2), "%")
# cm = confusion_matrix(y_test, y_pred)
# print(cm)
```

```python
import joblib

# Perform one-hot encoding on your dataset
X = pd.get_dummies(df.drop(['region'], axis=1), columns=columns_to_encode)

# Extract the list of categorical columns and the column order
categorical_columns = df.drop(['region'], axis=1).select_dtypes(include=['object']).columns.tolist()
column_order = X.columns.tolist()

# Create an encoding configuration dictionary
encoding_config = {
    'categorical_columns': categorical_columns,
    'column_order': column_order
}
```

```python
# Save the encoding configuration dictionary as a .pkl file
with open('encoding_config3.pkl', 'wb') as config_file:
    joblib.dump(encoding_config, config_file)
```
